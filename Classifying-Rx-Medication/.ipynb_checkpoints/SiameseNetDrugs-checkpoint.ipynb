```python
def create_pairs(x, digit_indices):
    """ Positive and negative pair creation.
        Alternates between positive and negative pairs.
    """
    pairs = []
    labels = []
    medclass = []
    n = min([len(digit_indices[d]) for d in range(10)]) - 1
    for d in range(10):
        for i in range(n):
            z1, z2 = digit_indices[d][i], digit_indices[d][i+1]
            pairs += [[x[z1], x[z2]]]
            inc = random.randrange(1, 10)
            dn = (d + inc) % 10
            z1, z2 = digit_indices[d][i], digit_indices[dn][i]
            pairs += [[x[z1], x[z2]]]
            labels += [1, 0]
    return np.array(pairs), np.array(labels)
```

```python
traindata = np.load('/content/drive/My Drive/Datasets/drugs/oneshottrain.npy')/255
trainlabels = np.load('/content/drive/My Drive/Datasets/drugs/oneshotlabels.npy')
digit_indices = [np.where(trainlabels == i)[0] for i in range(10)]

trainpairs, trainlabels = create_pairs(traindata, digit_indices)
trainpairs.shape
```
